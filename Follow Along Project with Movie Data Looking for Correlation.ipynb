```python
#importing the libraries as needed

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plots we will create.

# Read in the data

df = pd.read_csv(r'C:\Users\jackg\Downloads\movies.csv')
```

```python
#Looking at the data

df.head()
```

```python
#Looking to see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
# Change the data type of the budget and gross field to int64
df = df.dropna(subset=['budget', 'gross'])
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df.dtypes
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop any duplciates

df.head().drop_duplicates()
```

```python
df.head()
```

```python
# building a scatterplot with budget vs gross revenue

plt.scatter(x=df['budget'], y=df['gross'])


plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
#Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#start to look at the corelation

df.corr(method = 'pearson') #pearson, kendall, spearman
```

```python
#high corelation between budget and correlation
```

```python
correlation_matrix = df.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at Comapny

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
