# Time Series Analysis

Firstly we need to clean the data to be in the correct format and we need to extract information from the InvoiceDate column. 

```
df_19 = _deepnote_execute_sql('--Allows me to see the format of the Date. \nSELECT InvoiceDate\nFROM updated_data.csv\nLIMIT 1\n\n--I will need to work around this format to get the month and the actual day', 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_19
```

Used python to add the day, month, year columns to our data frame. Will use SQL below to analyze. 

```
import pandas as pd

df = pd.read_csv('updated_data.csv')

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%Y %H:%M')

df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Year'] = df['InvoiceDate'].dt.year

df.to_csv('updated_data_timeseries.csv', index=False)
df.head()
```
